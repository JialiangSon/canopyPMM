```python
import numpy as np
import matplotlib.pylab as plt
plt.rc('font',family='Times New Roman')

#####################################
plt.tick_params(top=True,right=True)
plt.rcParams['ytick.direction']='in'
plt.rcParams['xtick.direction']='in'
plt.tick_params(labelsize=14)
axes=plt.subplot()
axes.minorticks_on()
# plt.tick_params(which='major',width=2)
plt.tick_params(which='major',length=6)
# plt.tick_params(which='minor',width=2)
# plt.tick_params(which='minor',length=3)
#####################################

inU = np.loadtxt('PC/porousCylinder/InU/0/surfaceFieldValue.dat',skiprows=4)
inP = np.loadtxt('PC/porousCylinder/InP/0/surfaceFieldValue.dat',skiprows=4)
outP = np.loadtxt('PC/porousCylinder/OutP/0/surfaceFieldValue.dat',skiprows=4)
dp = inP[:,1]-outP[:,1]
print(dp)

# plt.plot(inU[:,0],inU[:,1],label='U')
plt.plot(inP[:,0],inP[:,1],linestyle='--',c='b',label='RANS-PMM_P$_{front}$')
plt.plot(outP[:,0],outP[:,1],linestyle='-.',c='r',label='RANS-PMM_P$_{back}$')
plt.plot(outP[:,0],dp,c='k',label='RANS-PMM_P$_{gradient}$',linestyle='--')
plt.xlim([0,300])

dpa = np.average(dp[237:312])
print(dpa)

##############################################

# inU = np.loadtxt('../porousCylinder/postProcessing/I/0/surfaceFieldValue.dat',skiprows=4)
inP = np.loadtxt('PC/multicylinder/inP_faceZoneAverage/0/surfaceFieldValue.dat',skiprows=4)
outP = np.loadtxt('PC/multicylinder/outP_faceZoneAverage/0/surfaceFieldValue.dat',skiprows=4)
dp = inP[:,1]-outP[:,1]
# print(dp)

# plt.plot(inU[:,0],inU[:,1],label='U')
plt.plot(inP[:,0],inP[:,1],linestyle='-',c='b',label='LES_P$_{front}$')
plt.plot(outP[:,0],outP[:,1],linestyle='-',c='r',label='LES_P$_{back}$')
plt.plot(outP[:,0],dp,c='k',label='LES_P$_{gradient}$',linestyle='-')

################################################
plt.legend()
plt.grid(linestyle='--')
plt.legend(fontsize=10)
plt.xlabel('Time',fontsize=16)
plt.ylabel('Pressure',fontsize=16)
plt.ylim([-1.2,1])

##############################################
```
